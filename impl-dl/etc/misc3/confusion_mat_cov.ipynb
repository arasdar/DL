```python
import numpy as np
```

```python
# Calculating the covariance matrix or cross correlation
X = np.random.rand(50, 1000)
```

```python
X.shape, X.dtype
```

```python
# cov = X @ X.T
cov = X.T @ X
```

```python
cov.shape

import matplotlib.pyplot as mplt
```

```python
mplt.imshow(cov)
mplt.show()
```

```python
X.shape, X.mean(axis=0).shape
X_norm = (X - X.mean(axis=0))/ X.std(axis=0)
```

```python
X_norm.shape
cov = X_norm.T @ X_norm
```

```python
mplt.imshow(cov)
mplt.show()
```
